```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
x=pd.read_csv("/content/my_movies.csv")
```

```python
x
```

```python
x.info()
```

```python
x.shape
```

```python
x2=x.iloc[:,5:]
x2
```

```python
frequent_itemsets=apriori(x2,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules[rules.lift>1]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
frequent_itemsets2=apriori(x2,min_support=0.05,use_colnames=True)
frequent_itemsets2
```

```python
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.9)
rules2
```

```python
rules2[rules2.lift>1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
